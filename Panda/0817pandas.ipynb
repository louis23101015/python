```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font',family="Malgun Gothic")
plt.rcParams['axes.unicode_minus'] = False
pd.read_csv("Koweps_hpc10_2015_beta1.csv")
```

```python
df = pd.read_csv("Koweps_hpc10_2015_beta1.csv")
codebook = pd.read_excel("Koweps_Codebook.xlsx")
```

```python
nc = dict(codebook[["변수명","설명"]].values)
df.rename(columns=nc,inplace=True)
val = list(nc.values())
df[val]
```

```python
# 성별 전처리
# 1 = 남  2 = 여  9 = Nan

# temp = {1:"남",2:"여"}

temp = {}
for i in codebook["내용"][0].split():
    a = i.split('.')
    temp[int(a[0])] = a[1]
    
df["성별"] = df.apply(lambda x: temp[x.성별] if x.성별 <= 2 else pd.NA, axis=1)
df['성별']
```

```python
# 0 = 비해당(18세 미만)  1 = 유배우  2 = 사별  3 = 이혼  4 = 별거  5 = 미혼(18세이상, 미혼모 포함)  6 = 기타(사망 등)

# temp2 = {0:"비해당(18세 미만)",
#          1:"유배우",
#          2:"사별",
#          3:"이혼",
#          4:"별거",
#          5:"미혼(18세 이상, 미혼모 포함)",
#          6:"기타(사망 등)"}

#temp = []
#temp2 = {}
#for i in codebook["내용"][2].split():
    #temp.extend(i.split("("))
#for i in temp:
    #if "." in i:
        #t = i.split(".")
        #temp2[int(t[0])] = t[1]

#df["혼인상태"] = df.apply(lambda x: temp2[x.혼인상태] if x.혼인상태 <= 6 else pd.NA, axis=1)
#df[val]


import re
marriage_data = codebook['내용'][2]
pattern = r'((\d+)\.(.*?)(?=\s\d+\.|$))' 
# (\d+)= \d: 숫자로시작, +:1개이상, (): 이 부분을 결과로 반환
# \.: 마침표찾기,
# (.*?) = .: 어떤 문자든 1개, *: 1개이상반복, ?: 가능한 짧게 
# # (?=\s\d+\.|$) = (?= ...): lookahead라고 불리며, ...에 들어간 패턴이 뒤 따라오는 경우에만을 의미
# # \s: 공백문자, \d+: 숫자하나이상, \.마침표, |: 또는 $: 문자열 끝
marriage_data = re.findall(pattern=pattern,string=marriage_data)
df['혼인상태'] = df["혼인상태"].map({int(k):v.strip() for s,k,v in marriage_data})
df[val]
```

```python
gmi = df.groupby(["성별","혼인상태"]).size().unstack("혼인상태")
gmi["총 인원"] = gmi.sum(axis=1)
gmi["혼인한 사람"] = gmi["총 인원"] - gmi["비해당(18세 미만)"] - gmi["미혼(18세이상, 미혼모 포함)"]
gmi["이혼율"] = (gmi["이혼"] / gmi['혼인한 사람'] * 100).round(2)
gmi
```

```python
#1. 태어난 연도 데이터를 가지고 2015년 기준으로 현재 나이를 나타내는 '나이'열 df에 추가하기
#2. 나이를 기준으로 0 ~ 29 청년 30 ~ 59 중년 60이상 노년 '연령대'열 추가
#3. 각 연령대 별 남/녀 성비 시각화(막대그래프: X 연령대 Y 인원수)
df['나이'] = 2015 - df['태어난 연도']

bins = [0,29,59,100]
labels = ['청년', '중년', '노년']
df['연령대'] = pd.cut(df['나이'], bins, labels=labels)

df[['나이','연령대']]
```

```python
age_gender_data = df.groupby(['연령대','성별']).size().reset_index(name='인원수')
 

sns.barplot(age_gender_data, x= '연령대', y = '인원수', hue = '성별')
plt.title("연령대 별 남여 성비 시각화")
plt.show
```

```python
#### 1. 남성과 여성의 임금차이 시각화(기준: 나이)
#### 2. 각 지역별 연령대 비율을 조사해 노인 인구가 높은 순으로 시각화
#### 3. 월 임금과 이혼율에 대한 상관관계 시각화

income_data = df.dropna(subset=['일한달의 월 평균 임금']).groupby(['성별','나이'])['일한달의 월 평균 임금'].mean().reset_index(name='연령별 평균 월급')
income_data
```

```python
income_data2 = {}
for gender in ['남','여']:
    temp = {}
    for age in range(100):
        cnt = 0
        total_income = 0
        for income in df[(df['나이'] == age) & (df['성별'] == gender)]['일한달의 월 평균 임금'].dropna():
            cnt += 1
            total_income += income
        if cnt:
            temp[age] = {'인원수':cnt, '총임금': round(total_income,2)}
            temp[age]['일한달의 월 평균 임금'] = round(total_income / cnt,2)
        income_data2[gender] = temp
    income_data2
```

```python
for gender in income_data2:
    for age in income_data2[gender]:
        a,b = income_data2[gender][age]['일한달의 월 평균 임금'] ,  round(income_data[(income_data['나이'] == age) & (income_data['성별'] == gender)]['연령별 평균 월급'].values[0],2)
        if a != b:
            print(a,b)
```

```python
sns.lineplot(income_data,x='나이',y='연령별 평균 월급', hue='성별')
plt.title("연령별 남여 월 평균 임금 차이")
plt.show()
```
